```python
# Import pandas to work with tabular (table-like) data
import pandas as pd
```

```python
# Google Sheets link exported as an Excel file
file_url = "https://docs.google.com/spreadsheets/d/1OGs8KoHGEZDxyFdaxAKmZbeO1BiSWVQ6/export?format=xlsx"
```

```python
# The actual column names start around row 20.
# We explicitly tell pandas to use row 20 as the header.
df = pd.read_excel(
    file_url,
    sheet_name="Canada by Citizenship",
    header=20
)
```

```python
# Display the first few rows to confirm the data loaded correctly
df.head()
```

```python
# Display column names to verify that expected columns exist
df.columns
```

```python
# Keep only:
# - Immigrants (Type == "Immigrants")
# - Foreigners (Coverage == "Foreigners")
df = df[
    (df["Type"] == "Immigrants") &
    (df["Coverage"] == "Foreigners")
]
```

```python
# Preview filtered dataset
df.head()
```

```python
# Rename technical UN column names to more readable names
df.rename(columns={
    "OdName": "Country",
    "AreaName": "Continent",
    "RegName": "Region",
    "DevName": "Development_Status"
}, inplace=True)
```

```python
# Check renamed columns
df.head()
```

```python
# Extract columns that represent years (e.g., 1980–2013)
year_columns = [col for col in df.columns if str(col).isdigit()]
```

```python
# Display year columns
year_columns[:10]
```

```python
# Convert year columns to numeric values
# - errors='coerce' converts invalid values to NaN
# - fillna(0) replaces missing values with 0
# - astype(int) converts values to integers
df[year_columns] = df[year_columns].apply(
    pd.to_numeric, errors="coerce"
).fillna(0).astype(int)
```

```python
# Check cleaned year data
df[year_columns].head()
```

```python
# Select only the required columns in the correct order
final_df = df[
    ["Country", "Continent", "Region", "Development_Status"] + year_columns
].copy()
```

```python
# Preview final structure
final_df.head()
```

```python
final_df.loc[:, "Total"] = final_df[year_columns].sum(axis=1)

final_df.head()
```
